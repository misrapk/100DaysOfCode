```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
x = iris.data[:, 2:] #petal length
y = iris.target
```

```python
#create tree classifieer
treeClf = DecisionTreeClassifier(max_depth=2)
treeClf.fit(x,y)
```

## Visualize the Tree


```python
from sklearn.tree import export_graphviz
```

```python
f = open("D:\\pkworkspace\\100DaysDataScience\\Step3 ML ALGO\\irisTree.png",'w')
export_graphviz(
    treeClf, out_file=f,
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True)
```
