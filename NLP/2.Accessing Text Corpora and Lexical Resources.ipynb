### 1.1 Gutenberg Corpus
it contains 25,000 free electronic books

```python
import nltk
```

```python
from nltk.corpus import gutenberg
gutenberg.fileids()
```

```python
#lets pick one
emma = nltk.corpus.gutenberg.words('austen-emma.txt')
len(emma)
```

```python
# lets display other info of gutenberg texts
# avg word length, avg sentence length, number of times each vocab appear in text on avg
for fileid in gutenberg.fileids():
    numChars = len(gutenberg.raw(fileid))
    numWords = len(gutenberg.words(fileid))
    numSents = len(gutenberg.sents(fileid))
    numVocab = len(set([w.lower() for w in gutenberg.words(fileid)]))
    
    print(int(numChars/numWords), int(numWords/numSents), int(numWords/numVocab), fileid)
```

### 1.2 Brown Corpus
1. First million word electronic Corpus of English
2. created in 1961 at Brown university.
3. Categorised by Genre

```python
from nltk.corpus import brown
```

```python
brown.categories()
```

```python
brown.words(categories='news')
```

```python
#lets compare genres in their usage of modal verbs
```

```python
#step1 : Produce counts for particular genre
learnedText = brown.words(categories='learned')
fdist = nltk.FreqDist([w.lower() for w in learnedText])
modals = ['what', 'when', 'who', 'where','why']

for m in modals:
    print('{} : {}'.format(m, fdist[m]))
```

```python
# step 2: obtain counts for each genre of intrest.
```

```python
cfd = nltk.ConditionalFreqDist(
            (genre, word)
            for genre in brown.categories() 
            for word in brown.words(categories=genre))
genres = ['adventure',
 'belles_lettres',
 'editorial',
 'fiction',
 'government',
 'hobbies',
 'humor',
 'learned',
 'lore',
 'mystery',
 'news',
 'religion',
 'reviews',
 'romance',
 'science_fiction']
models = ['what', 'when', 'who','where','why','nude']
cfd.tabulate(conditions=genres, samples = models)
```

### 1.4 Inaugural Address Corpus
1. collection of 55 texts, one for each predsidential adddress.

```python
from nltk.corpus import inaugural
inaugural.fileids() 
```

```python
[fileid[:4] for fileid in inaugural.fileids()]
```

```python
cfd = nltk.ConditionalFreqDist( (target, fileid[:4])
                               for fileid in inaugural.fileids()
                               for w in inaugural.words(fileid)
                               for target in ['america','citizen']
                               if w.lower().startswith(target))
cfd.plot()
```

#### udhr --> Universal Declaration of Human Rights in 300 lang.


```python
from nltk.corpus import udhr
languages = ['Chickasaw', 'English', 'German_Deutsch','Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']
cfd = nltk.ConditionalFreqDist((lang,len(word))
                              for lang in languages
                              for word in udhr.words(lang + '-Latin1'))
cfd.plot(cumulative = True)
```

```python
raw_text = udhr.raw('Hindi-UTF8')
nltk.FreqDist(raw_text).plot()
```
