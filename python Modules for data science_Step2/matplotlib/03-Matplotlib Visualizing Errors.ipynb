In Visualization of data and results, showing error effectively can make a plot convey much more complete information.

# Basic Errorbars

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0,10,50)
dy = 0.8   #error
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt ='.k')
```

```python
# many more properties of errorbar
plt.errorbar(x,y, yerr= dy, fmt='o',color='black',
            ecolor = 'lightgray', elinewidth = 3, capsize=4)
```
