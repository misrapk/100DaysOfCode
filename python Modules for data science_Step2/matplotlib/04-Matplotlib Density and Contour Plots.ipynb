# Density and Contour Plots
For displaying 3D data in 2D dimensions using Contour region. 

funcitons: 
           1. plt.contour ---> contour plots
           2. plt.contourf ---> filled contour plots
           3. plt.imshow ---> showing images

```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

### Visualizing a 3-D funciton
Demonstarte using z = f(x,y)

```python
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10 + y*x) * np.cos(x)
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
#plt.contour(x,y,z)   x and y position on plot and z represent the contour level

plt.contour(X,Y,Z, colors = 'black')


#-ve value represented by (-) and positive is by solid lines
```

```python
# draw more lines within data range
plt.contour(X,Y,Z, 20, cmap='RdGy')  #RdGy --> Red-Gray
```

```python
# plt.cm.<TAB>      #for knowing the color option
```

```python
# filling the gapped line between the plot
plt.contourf(X,Y,Z, 20, cmap = 'RdGy')
plt.colorbar()
```
