```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
#lets plot trignometric spiral

ax = plt.axes(projection='3d')


#data
zline = np.linspace(0,15,1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline)

#data for scatter points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata)+0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c = zdata, cmap='Reds')
```

### 3D Countour plots

using ax.contour3D

```python
def f(x,y):
    return np.sin(np.sqrt(x ** 2 + y**2))
```

```python
x = np.linspace (-6,6,30)
y = np.linspace(-6,6,30)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z, 50, cmap = 'Blues')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
# to view on different angle
ax.view_init(60, 40)    #ax.view_init(angle, roation along z)
fig
```

### Wireframes and Surface Plots

Take the grid of values and project it onto the specified 3D surface

```python
#wireframe

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z, color='black')
ax.set_title('wireframe')
```

```python
ax.view_init(60,40)
fig
```

```python
# surface plot --> add the color to the polygon
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z, rstride = 1, cstride=1,
               cmap= 'viridis', edgecolor='none')
ax.set_title('surface')

#change the rstride(row stride) and cstride(column stride) to make it interesting
```
