# Histogram

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
data = np.random.randn(1000)
plt.hist(data)
```

```python
#more cutomized

plt.hist(data, bins=30, density=True, alpha=0.5,histtype='stepfilled',
        color = 'steelblue', edgecolor='black')
```

```python
#multiple histogram
x1= np.random.normal(0, 0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)
```

```python
kwargs = dict(histtype='stepfilled',alpha=0.5, density=True, bins = 40)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

# 2D histogram an binnings

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean, cov, 10000).T
```

### plt.hist2d: 2D histogram

```python
plt.hist2d(x,y, bins=30, cmap='Reds')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

#### plt.hexbin --> hexagonal representation

```python
plt.hexbin(x, y, gridsize=40, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```
