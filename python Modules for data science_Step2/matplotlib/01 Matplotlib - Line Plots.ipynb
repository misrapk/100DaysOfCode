```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
#create fig and axes
fig = plt.figure()
ax = plt.axes()
```

```python
# use ax.plot to plot any data


x = np.linspace(0,10, 10000)
ax.plot(x, np.sin(x))
```

```python
# we can also do like
plt.plot(x, np.sin(x))
```

```python
# if u want to create single fig multiple times
# call plot function multiple times
plt.plot(x,np.sin(x))
plt.plot(x, np.cos(x))
# plt.plot(x, np.tan(x))
```

### Adjusting Line Color and Styles

```python
plt.plot(x, np.sin(x-0), color = 'blue')     #color name
plt.plot(x, np.sin(x-1), color = 'g')        #short name
plt.plot(x, np.sin(x-2), color = '0.55')       #Grayscale btwn 0 and 1
plt.plot(x, np.sin(x-3), color = '#FFDD22')    #HEX code(RRGGBB)
plt.plot(x, np.sin(x-5), color = (1.0, 0.2, 0.3))  #RGB tuple   
plt.plot(x, np.sin(x-7), color = 'chartreuse')  #all Html color names supported
```

```python
# Adjust line style using linestyle keyword
plt.plot(x, np.cos(x-0), linestyle='solid')     #'-'
plt.plot(x, np.cos(x-2), linestyle='dashed')    # '--'
plt.plot(x, np.cos(x-4), linestyle='dashdot')   # '-.'
plt.plot(x, np.cos(x-5), linestyle='dotted')    # ':'

```

```python
#combine both thing in one line
plt.plot(x, x*1, '-g')   #solid green
plt.plot(x, x*3, '--b')   #dashed blue
plt.plot(x, x*5, '-.k')   #dashdot black
plt.plot(x, x*7, ':r')   #dotted red
plt.plot(x, x*9, ':y')   #solid green
plt.plot(x, x*11, '-b')   #solid green

```

### Adjusting the Axes limits

use plt.xlim() and plt.ylim()

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.2,1.2)
```

```python
#display the axis in reverse
plt.plot(x, np.sin(x))
plt.xlim(11,-1)
plt.ylim(1.5,-1.5)
```

plt.axis() allows you to set the x and y limits

[xmin, xmax, ymin, ymax]

```python
plt.plot(x, np.cos(x))
plt.axis([-1,11, -1.5, 1.5])
```

```python
# plt.axis() also do automatically
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
plt.plot(x, np.tan(x))
plt.axis('tight')
```

### Labelling plots

```python
#using axes

ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0,10), ylim=(-2,2),
       xlabel='x', ylabel='sin(x)',
       title='SinGraph')
plt.show()
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend();
```
