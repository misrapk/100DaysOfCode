```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

#### Scatter plot with plt.plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)

#plt.plot(x,y, charac that represent symbol, color)
plt.plot(x,y, 'o', color = 'black')
```

```python
# We can use so many symbols
rng = np.random.RandomState(0)
for marker in ['o','.',',','x','+','v','^','<','>','s','d']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
            label = "marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0,1.8)
```

```python
#we can use character with the line
plt.plot(x,y, '-ok')    #line(-), circle(o), black(k)
```

```python
#also
plt.plot(x, y, '-p', color='gray',
        markersize = 15, linewidth=4,
        markerfacecolor = 'white',
        markeredgecolor = 'gray',
        markeredgewidth = 2)
plt.ylim(-1.2, 1.2)
```

#### Scatter with plt.scatter

```python
# y = np.cos(x)
plt.scatter(x, y , marker='o')
```

```python
#create random plots to show the properties
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
```

```python
plt.scatter(x, y, c=colors, s=sizes, alpha = 0.3,
           cmap = 'viridis')
plt.colorbar()
```

## Example: WOrk on IRIS dataset

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
features = iris.data.T
```

```python
plt.scatter(features[0], features[1], alpha=0.2,
           s = 100*features[3], c = iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
