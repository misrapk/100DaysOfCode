Salary is as object type so first we have to convert it into numeric type

```python
def cleanupSalary(row):
    salary = row['Salary'].replace('$', '')
    salary = float(salary)    #type casting
    return salary
```

```python
df['cleanSalary'] = df.apply(cleanupSalary, axis = 1)
```

```python
df.head()
```

```python
#now take mean of clean salary
print("Average Salary per gender")
df.groupby('gender')['cleanSalary'].mean()
```

# Task 2
Group by the "Job Title" and avg salary of female and male

For this create new separate column for female salary and male salary and take mean of them with grouping te job title

```python
def fSalary(row):
    if row['gender'] == 'Female':
        return row['cleanSalary']
df['femaleSalaryavg'] = df.apply(fSalary, axis =1)
df.head()
```

```python
def mSalary(row):
    if row['gender'] == 'Male':
        return row['cleanSalary']
df['MaleSalaryavg'] = df.apply(mSalary, axis =1)
df.head()
```

```python
#Group by with job tile
df.groupby('Job Title')['cleanSalary', 'MaleSalaryavg', 'femaleSalaryavg'].mean()
```

# TASK 3

```python
#Group by with job tile
df.groupby(['Job Title','City'])['cleanSalary', 'MaleSalaryavg', 'femaleSalaryavg'].mean()
```

## Task 2

```python
#create male column
def mByte(row):
    if row['gender'] == 'Male':
        return row['bytes']
    
#create female column
def fByte(row):
    if row['gender'] == 'Female':
        return row['bytes']
```

```python
df2['maleBytes'] = df2.apply(mByte, axis=1)
df2['femaleBytes'] = df2.apply(fByte, axis=1)
```

```python
df2.head()
```

```python
#groupby total byte
df3=df2.groupby('url')['maleBytes', 'femaleBytes'].sum()
df3
```

```python
df2.groupby('gender')['bytes'].sum()
```

```python
df4 = df2.groupby(['url', 'country'])['femaleBytes', 'maleBytes'].sum()
df4.head()
```
